한국의 데이터를 주로 하되, 미국의 기준금리도 중요한 feature이기에 넣어준다.

```python
date_index = kor_df.index
df = kor_df.copy()
df['미국_기준금리(연%)'] = us_df.loc[date_index, 'funds_rate']
df.head()
```

소비자물가지수 -> 소비자물가상승률(consumer inflation rate)<br>
보통 물가상승률은 전년동월대비 상승률로 표시한다.<br>
물가상승률 = (금월 물가지수 - 전년동월 물가지수) / 전년동월 물가지수 * 100

```python
df_added = df.copy()
date_index = df.index
cpi_col = '소비자물가지수_총(2020=100)'
for i in range(len(date_index)):
  if i not in [0, 11]:  # 처음 년도 인덱스는 제외
    present_date_i = df.index[i]
    past_date_i = df.index[i-12]
    present = df.loc[present_date_i, cpi_col]
    past = df.loc[past_date_i, cpi_col]
    temp = (present - past) / past * 100  # 물가상승률
    df_added.loc[present_date_i, cpi_col] = temp
df_added.drop(df_added.index[0:12], inplace=True)
```

```python
df_added.rename(columns={cpi_col:'소비자물가상승률(%)'}, inplace=True);
df_added.head()
```

시계열 데이터이므로, 시계열 분석에 맞는 모델을 만들어야 한다.<br>
먼저 정상성을 지닌 시계열 데이터로 바꾸고자 한다.(정상성 체크 by ADF test)

```python
df = df_added.copy()  # 다시 대체해준다
```

```python
# from statsmodels.tsa.stattools import adfuller
def adf_test(y):
  dftest = adfuller(y, autolag='AIC')
  pvalue = dftest[1]
  return pvalue
unsatisfied_cols = []
for col in df.columns:
  print(f"{col}: {adf_test(df[col]).round(4)}")
print()
for col in df.columns:
  print(f"{col}: {adf_test(df[col])<0.05}") # True면 정상성 만족
  if adf_test(df[col]) >= 0.05: # unsatisfied
    unsatisfied_cols.append(col)
```

환율, 경제성장률, 실업률, 미국 기준금리 제외하고 모두 정상성을 띠지 않는다.

```python
unsatisfied_cols  # 정상성을 만족하지 않는 컬럼들
```

unsatisfied_cols는 차분해준다.(금월 값 - 전월 값)(변화량)

```python
date_index = df.index
df_stat = df.copy() # 정상성 띠도록 만든 데이터(stationary)

for i in range(len(date_index)):
  if i != 0:  #i==0일 때는 변화량 계산 불가
    present_date_i = df.index[i]
    past_date_i = df.index[i-1]
    present = df.loc[present_date_i, unsatisfied_cols]
    past = df.loc[past_date_i, unsatisfied_cols]
    temp = present - past # 변화량
    df_stat.loc[present_date_i, unsatisfied_cols] = temp
```

```python
for col_i in range(len(df_stat.columns)):
  print(f"{df_stat.columns[col_i]}: {adf_test(df_stat.iloc[1:, col_i])<0.05}") # 0번째 인덱스는 변화량이 되지 않았으므로 제외함
```

외환보유액_금(천달러)만 정상성 만족하지 않음.<br>
이 컬럼은 2차 차분해준다.(변화량의 변화량)

```python
df_stat.drop(df_stat.index[0], inplace=True)  # 0번째 인덱스는 제거해준다
```

```python
date_index = df_stat.index
df_stat_second = df_stat.copy()
second_unsatisfied_cols = ['외환보유액_금(천달러)']
for i in range(len(date_index)):
  if i != 0:  # 0번째면 제외
    present_date_i = df_stat.index[i]
    past_date_i = df_stat.index[i-1]
    present = df_stat.loc[present_date_i, second_unsatisfied_cols]
    past = df_stat.loc[past_date_i, second_unsatisfied_cols]
    temp = present - past
    df_stat_second.loc[present_date_i, second_unsatisfied_cols] = temp
```

```python
for col_i in range(len(df_stat_second.columns)):
  print(f"{df_stat_second.columns[col_i]}: {adf_test(df_stat_second.iloc[1:, col_i])<0.05}") # 0번째 인덱스는 변화량이 되지 않았으므로 제외함
```

이제 모든 컬럼이 정상성을 띠게 되었다.(df_stat_second)
