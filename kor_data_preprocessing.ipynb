# **경제성장률**
- 분기별 데이터(1960/Q2~2022/Q2) -> 월별 데이터로 upsampling하기

물론 데이터를 인위적으로 늘리는 만큼 정확도는 떨어지지만, 경제성장률은 무시하기엔 중요한 feature임.

```python
# files.upload(); # 분기별_경제성장률.csv
```

```python
url = '분기별_경제성장률.csv'
growth_rate = pd.read_csv(url).T
```

```python
growth_rate.head()
```

```python
growth_rate.columns = ['경제성장률']
growth_rate = growth_rate.iloc[4:,:]
growth_rate.head()
```

이 분기별 데이터를 월별 데이터로 바꾸고자 한다(upsampling)<br>
먼저 datetime index로 바꿔준다.

```python
monthly_growth_rate = growth_rate.copy()
monthly_growth_rate.head()
```

Q1 ~ Q4는 컴퓨터가 이해하기 어려우므로 각각 2월, 5월, 8월, 11월로 바꿔준다<br>
(1분기는 1 ~ 3월인데, 분기초로 잡으면 해석이 잘못될 수 있으므로 분기의 중간으로 쓴다)

```python
date_index = monthly_growth_rate.index
date_index
```

```python
date_index = date_index.str.replace('Q1', '02')
date_index = date_index.str.replace('Q2', '05')
date_index = date_index.str.replace('Q3', '08')
date_index = date_index.str.replace('Q4', '11')
date_index
```

```python
monthly_growth_rate.index = pd.to_datetime(date_index)
monthly_growth_rate.head()
```

```python
monthly_growth_rate = monthly_growth_rate.resample('MS').mean()
monthly_growth_rate
```

```python
# 결측치 채우기
monthly_growth_rate = monthly_growth_rate.interpolate(method='linear')  # 월마다 30, 31일로 다 달라서 동일 간격이 아닐 수 있으므로 선형으로 보간
monthly_growth_rate
```
